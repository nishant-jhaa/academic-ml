```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
kd = pd.read_csv('data.csv', delimiter=',')
kd = kd.drop(list(kd)[4:-2],axis=1)
kd = kd.drop(list(kd)[0],axis=1)
kd = np.array(kd)

te = TransactionEncoder()
te_ary = te.fit(kd).transform(kd)
df = pd.DataFrame(te_ary,columns=te.columns_)
fi = apriori(df,min_support=0.1,use_colnames=True)

# this part is for apriory all for the item combination above 2 itemsets and minimum support 0.1
fi[fi['itemsets'].apply(lambda x:len(x))>2]
```

```python
# correlation part for showing confidence and lift

# association rules created with rules having lift more than 1
rules = association_rules(fi, metric="lift", min_threshold=1)
xx = rules['antecedents'].apply(lambda x:len(x))
final_rules = rules[xx>2].sort_values('confidence',ascending = False)

# choosing confidence more than 0.8
final_rules[final_rules['confidence']>.8]
```

```python
#label encoding for model training
from sklearn.preprocessing import LabelEncoder
Features = rd.drop('gender',axis=1)
Target = rd['gender']
label = LabelEncoder()
Cat_Colums = Features.dtypes.pipe(lambda Features: Features[Features=='object']).index
for col in Cat_Colums:
    Features[col] = label.fit_transform(Features[col])
```
